```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
def opPerceptron(X, y, lr=0.1, epochs=1000):
    np.random.seed(42) 
    w1, w2 = np.random.randn(2)
    b = 0

    for _ in range(epochs):
        for i in range(X.shape[0]):
            z = w1 * X[i][0] + w2 * X[i][1] + b
            if z * y[i] <= 0:
                w1 += lr * y[i] * X[i][0]
                w2 += lr * y[i] * X[i][1]
                b += lr * y[i]

    return w1, w2, b            
```

```python
X, y  = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
w1, w2, bias = opPerceptron(X,y,lr=0.1,epochs=1000)
```

```python
slope = -(w1/w2)
intercept = -(bias/w2)
```

```python
print(f"Weight1: {w1}")
print(f"Weight2: {w2}")
print(f"Bias: {bias}")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

```python
x_input = np.linspace(-3,3,100)
y_input = slope * x_input + intercept
```

```python
plt.plot(x_input,y_input,color='red')
plt.scatter(X[:,0],X[:,1],c=y)
plt.ylim(-2,2)
plt.show()
```
